## MIDO LAS MEDIAS REALES DE LOS 3 DF PARA VER QUE TANTO SESGO TENDRAN MIS MODELOS

```python
target_0= df_0['product']
target_mean_0= target_0.mean()
print("Valor promedio real del target 0:", target_mean_0)


target_1= df_1['product']
target_mean_1= target_1.mean()
print("Valor promedio real del target 1:", target_mean_1)


target_2= df_2['product']
target_mean_2 = target_2.mean()
print("Valor promedio real del target 2:", target_mean_2)
```

### Comienzo a preparar el modelo para la primera base de datos (df_0)

```python
#Modelo de regresión lineal para del DF 0
#Separo los datos por las caracteristicas y los objetivos a encontrar
features_0=df_0.drop(['id','product'],axis=1)
target_0 = df_0['product']

#Separo la cantidad de datos, 75% para entrenamiento y 25% para validación 
features_train_0,features_valid_0, target_train_0,target_valid_0 = train_test_split(
    features_0,
    target_0,
    test_size=0.25,
    random_state=12345)

#Comienzo a crear el modelo
model_0=LinearRegression()
model_0.fit(features_train_0,target_train_0)
predictions_valid_0=model_0.predict(features_valid_0)
mean_pred_0 = predictions_valid_0.mean()
result_0=mean_squared_error(target_valid_0,predictions_valid_0)**0.5
print('RMSE:', result_0)
print('PROMEDIO:', mean_pred_0)
```

```python
#Modelo de regresión lineal para del DF 1
#Separo los datos por las caracteristicas y los objetivos a encontrar
features_1=df_1.drop(['id','product'],axis=1)
target_1 = df_1['product']

#Separo la cantidad de datos, 75% para entrenamiento y 25% para validación 
features_train_1,features_valid_1, target_train_1,target_valid_1 = train_test_split(
    features_1,
    target_1,
    test_size=0.25,
    random_state=12345)

#Comienzo a crear el modelo
model_1=LinearRegression()
model_1.fit(features_train_1,target_train_1)
predictions_valid_1=model_1.predict(features_valid_1)
mean_pred_1 = predictions_valid_1.mean()
result_1=mean_squared_error(target_valid_1,predictions_valid_1)**0.5
print('RMSE:', result_1)
print('PROMEDIO:', mean_pred_1)
```

```python
#Modelo de regresión lineal para del DF 2
#Separo los datos por las caracteristicas y los objetivos a encontrar
features_2=df_2.drop(['id','product'],axis=1)
target_2 = df_2['product']

#Separo la cantidad de datos, 75% para entrenamiento y 25% para validación 
features_train_2,features_valid_2, target_train_2,target_valid_2 = train_test_split(
    features_2,
    target_2,
    test_size=0.25,
    random_state=12345)

#Comienzo a crear el modelo
model_2=LinearRegression()
model_2.fit(features_train_2,target_train_2)
predictions_valid_2=model_2.predict(features_valid_2)
mean_pred_2 = predictions_valid_2.mean()
result_2=mean_squared_error(target_valid_2,predictions_valid_2)**0.5
print('RMSE:', result_2)
print('PROMEDIO:', mean_pred_2)
```

## Prepárate para el cálculo de ganancias:
### Almacena todos los valores necesarios para los cálculos en variables separadas.

### Dada la inversión de 100 millones por 200 pozos petrolíferos, de media un pozo petrolífero debe producir al menos un valor de 500,000 dólares en unidades para evitar pérdidas (esto es equivalente a 111.1 unidades). Compara esta cantidad con la cantidad media de reservas en cada región.

### Presenta conclusiones sobre cómo preparar el paso para calcular el beneficio.

```python
#Establecer las variables a utilizar
threshold = 111.1
n_top = 200
earns = 4500
```

```python
#Comparamos la media obtenida en el modelo anterior con la media esperada "111.1 unidades"
print('                 REGION 0        ')
print()

if mean_pred_0 >= threshold:
    print("La región cumple con el umbral de rentabilidad.")
else:
    print("La región NO cumple con el umbral de rentabilidad.")

#Paso 4: Calcular ganancia potencial
# Seleccionar los 200 pozos con las predicciones más altas.


sorted_preds = np.sort(predictions_valid_0)
if len(sorted_preds) >= n_top:
    top200 = sorted_preds[-n_top:]
else:
    top200 = sorted_preds

# Sumar las reservas de los 200 mejores pozos
total_reserve = top200.sum()
# Calcular la ganancia potencial:

gain = (total_reserve - n_top * threshold) * earns
print()
print("Total reservas de los 200 mejores pozos:", total_reserve)
print()
print("Ganancia potencial total:", gain)
print()

# Paso 5: Evaluar riesgos y ganancias mediante bootstrapping 
n_iterations = 1000
boot_gains = []

# Realizar 1000 muestras (remuestreos) sobre los top200
for i in range(n_iterations):
    sample = np.random.choice(top200, size=len(top200), replace=True)
    sample_gain = (sample.sum() - n_top * threshold) * earns
    boot_gains.append(sample_gain)

boot_gains = np.array(boot_gains)
mean_gain = boot_gains.mean()
ci_lower = np.percentile(boot_gains, 2.5)
ci_upper = np.percentile(boot_gains, 97.5)
risk = (boot_gains < 0).mean() * 100

print("Ganancia promedio (bootstrapping):", mean_gain)
print()
print("Intervalo de confianza del 95%:", (ci_lower, ci_upper))
print()
print("Riesgo de pérdidas (%):", risk, "%")

```
